```python
#import the required library
from bs4 import BeautifulSoup
```

```python
#import webscrapping example html file
HTMLfilepath = "index.html" #change as per the path of your html file
with open(HTMLfilepath, "r") as organization:
    soup = BeautifulSoup(organization, "lxml")
```

```python
#view the content of the soup object
soup.contents
```

```python
#search using find method
tag_li = soup.find("li")
```

```python
#print tag type
print(type(tag_li))
```

```python
#print tag
print(tag_li)
```

```python
#search the document using find method for an ID
find_id = soup.find(id ='page-top')
```

```python
#print the find id object
print(find_id)
```

```python
#print string value
print(find_id.div.li.string)
```

```python
#search using string only
search_for_stringOnly = soup.findAll(text=["About"])
```

```python
#print the search value
print(search_for_stringOnly)
```

```python
#search based on css class name(presents as an attribute)
css_class_search = soup.find(attrs= {"class": "container-fluid p-0"})
print(css_class_search)
```

```python
#create a function to search the document based upon the tag passed as parameter
def is_account_manager(tag):
    return tag.has_attr("id") and tag.get("id") == "experience"
```

```python
#search the document using function and print it
account_manager = soup.find(is_account_manager)
print(account_manager.div.h2.string)
```

```python
#print tag name using True - which returns all the tags present in the document
for tag in soup.findAll(True):
    print(tag.name)
```

```python
#search using findall method for the given class
find_class = soup.findAll(class_ ='container-fluid p-0')
```

```python
#view the type of the class
type(find_class)
```

```python
#print the second resultset
print(find_class[0])
```

```python
#print the second result
#print(find_class[2])
```

```python
#find parents using parent method
find_class = find_class[0]
find_parent = find_class.find_parents("span")
print(find_parent)
```

```python
#now the find method to search based on the id
org = soup.find(id = "page-top")
```

```python
#print the search object
print(org)
```

```python
#find the next siblings
next_siblings = org.findNextSiblings()
```

```python
#print parents
parent = org.findParents
print(parent)
```

```python
#find and print previous
all_previous = org.findAllPrevious()
print(all_previous)
```

```python
#search and print previous sibling
previous_sibling = org.findPreviousSibling()
```

```python
#search and print all next
all_next = org.findAllNext()
print(all_next)
```

```python
#use regular expression to search the document
import re
email_example = """<br/>
<p> my email id is</p>
abc@example.com
"""
soup_email = BeautifulSoup(email_example, "lxml")
#use compile method to compile the information which contains regular expression
emailID_regexp = re.compile("\w+@\w+\.\w+")
#find and print the email id using regular expression
email_id = soup_email.find(text=emailID_regexp)
print(email_id)
```
